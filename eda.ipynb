```python
#Import necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset

df_titanic = pd.read_csv("C:/Users/user/Desktop/Prodigy Infotech_DS_02/train.csv")
```

```python
# Display the first few rows of the dataset

df_titanic.head()
```

```python
# Display the last few rows of the dataset

df_titanic.tail()
```

```python
# Get an overview of the dataset

df_titanic.info()
```

```python
# Get the mathematical overview of the dataset

df_titanic.describe()
```

```python
# Check for missing or null values in the dataset

df_titanic.isnull().sum()
```

```python
# Get the count of unique values for each column

df_titanic.nunique()
```

```python
# Display the shape of the dataset


df_titanic.shape
```

```python
# Fill missing Age values with the median age

df_titanic.fillna({"Age": df_titanic["Age"].median()}, inplace = True)
```

```python
# Confirm that Age missing values are filled

df_titanic.tail()

df_titanic.isnull().sum()
```

```python
# Drop irrelevant columns and prepare the dataset for analysis


df_titanic1= df_titanic.drop(["PassengerId", "Name", "Cabin", "SibSp", "Ticket"], axis= 1)
```

```python
# Checking the first 10 rows of the dataset to confirm that the irrelevant columns have been romoved

df_titanic1.head(10)
```

```python
# Convert Sex to binary: 0 for male, 1 for female

df_new_titanic = df_titanic1["Sex"].apply(lambda x: 0 if x == "male" else 1)

df_new_titanic.head(10)
```

```python
df_titanic1["Sex"] = df_new_titanic
```

```python
df_titanic1
```

```python
# Check for missing values after the conversion

df_titanic1.isnull().sum()
```

```python
# Fill missing Embarked values with the mode of the column

df_titanic1.fillna({"Embarked": df_titanic1["Embarked"].mode()[0]}, inplace = True)
```

```python
df_titanic1
```

```python
df_titanic1.isnull().sum()
```

```python
# Convert Embarked to numerical values: S=0, C=1, Q=2

df_titanic1['Embarked'] = df_titanic1['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# Calculating the correlation of thje dataset

df_titanic1.corr()
```

```python
# Calculate and plot the correlation heatmap of the dataset

plt.figure(figsize=(12, 6))
sns.heatmap(df_titanic1.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Heatmap of Titanic dataset showing Correlation")
plt.show()
```

```python
# Plot survival rate by Passenger Class

plt.figure(figsize=(8, 6))

sns.countplot(x='Pclass', hue='Survived', data=df_titanic1, palette='pastel')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Plot survival rate by Gender

plt.figure(figsize=(8,6))

sns.countplot(data = df_titanic1, x= "Sex", hue= "Survived", palette= "deep")
plt.title("Survival rate by Gender")
plt.show()
```

```python
# Plot survival rate by Embarked Location

plt.figure(figsize=(8,6))

sns.countplot(data = df_titanic1, x= "Embarked", hue= "Survived", palette= "Dark2")
plt.title("Survival rate by Embarked Location")
plt.show()
```

```python
# Plot survival rate by Parch

plt.figure(figsize=(8,6))

sns.countplot(data = df_titanic1, x= "Parch", hue= "Survived", palette= "coolwarm")
plt.title("Survival rate by Parch")
plt.show()
```

```python
# Plot Age distribution by Survival status

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df_titanic1[df_titanic1['Survived'] == 1], x='Age', fill=True, label='Survived')
sns.kdeplot(data=df_titanic1[df_titanic1['Survived'] == 0], x='Age', fill=True, label='Not Survived')
plt.title('Age Distribution by Survival')
plt.legend()
plt.show()
```

```python
# Plot Fare distribution by Survival status

plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=df_titanic1, hue="Survived", palette="coolwarm")
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
# Plot Survival Rate by Passenger Class and Gender

plt.figure(figsize=(8, 6))
sns.catplot(x='Pclass', hue='Survived', col='Sex', kind='count', data=df_titanic1, palette='husl')
plt.suptitle('Survival Rate by Passenger Class and Gender')
plt.show()
```

```python
# Plot Age vs Fare by Survival status

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', style='Survived', data=df_titanic1, palette='Set1')
plt.title('Age vs Fare by Survival')
plt.show()
```

```python
# Plot Pairplot of all numerical variables colored by Survival status

plt.figure(figsize=(20, 10))
sns.pairplot(data = df_titanic1, hue='Survived', palette="coolwarm")
plt.suptitle('Pairplot of All Variables', y= 1.02)
plt.show()
```
